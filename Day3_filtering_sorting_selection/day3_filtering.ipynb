column.

2.Why use it? Useful for finding titles (Mr., Mrs., Dr.) or specific
brands in a list of names.

Pro Tip: Use na=False to ignore rows where the name is missing

```python
# targ=['Countess', 'Col']

# data[data['Name'].str.contains(targ, na=False)]

# .str.contains() expects: A SINGLE STRING or (REGEX pattern -> Countess|Col)
```

```python
spec_ppl=data[data['Name'].str.contains('Countess|Col', na=False)]
spec_ppl
```

## Categorical/text columns:

MODE = most frequent value

we cannot use:

mean
median

because those work only for numbers.

```python
data['Embarked']=data['Embarked'].fillna('S')
```

```python
data['Embarked'].isnull().sum()
```

## 🧠 In Titanic dataset usually:

Family size is CREATED using:

SibSp + Parch + 1

Where:

SibSp = siblings/spouse

Parch = parents/children

+1 = passenger himself

```python
#To create a family size in titanic we use this
# Create (or overwrite) a column called family_size
data['family_size']= data['SibSp'] + data['Parch'] + 1
data['family_size']
```

```python
#Create Fare_Per_Person
# Create (or overwrite) a column called Fare_Per_Person
data['fare_per_person']=data['Fare']/ data['family_size']
data['fare_per_person']
```

```python
data[['Fare', 'family_size', 'fare_per_person']].head()
```

## Mini Analysis Questions

```python
#  1. Show survived females
survived_female=data[(data['Survived']==1) & (data['Sex']=='female')]

# 2. Show passengers younger than 18
under_18=data.query('Age < 18')
# If Age has missing values (NaN): NaN < 18 = False

# 3. Show passengers from class 1 with fare > 50
class1= data[(data['Pclass']==1) & (data['Fare']> 50)]

# 4. Show passengers whose names contain "Mr"
Mr_pass=data[data['Name'].str.contains('Mr', na=False)]

# 5. Show passengers who embarked from S
embark_s= data[data['Embarked']=='S']


survived_female.head()
```

## GroupBy (The Aggregator)
What is it? It collapses your data into groups based on a category.

The Process: Split (group by class) Apply (calculate average
age) Combine (show result).

Why use it? To compare performance or survival across different
demographics.

```python
data.groupby('Pclass')['Survived'].mean()
```
