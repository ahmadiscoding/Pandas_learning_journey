```python
import pandas as pd
import numpy as np
```

## What is a Series?

A Pandas Series:
- is one-dimensional (like a list)
- has labels called an index
- can store different types of data

Think of it like a single column in Excel.

## Creating Series from List
A Series is a 1D labeled array.

When we pass a Python list to pd.Series(), Pandas automatically assigns an index starting from 0.

```python
s=pd.Series([10,20,30])
s
```

We can access values using the index number.

```python
s[2]
```

## Creating Series with Custom Index
We can access values using the index number.

```python
s=pd.Series([101,104,108], index=["a","b","c"])
s
```

```python
s=pd.Series([101,104,108], index=["a","b","c"])
s.loc["c"]=5
s.loc["a"]=200
s
```
