# Day 1 - Pandas Foundations
Learning Series and DataFrame basics.

```python
import pandas as pd
import numpy as np
```

## What is a Series?

A Pandas Series:
- is one-dimensional (like a list)
- has labels called an index
- can store different types of data

Think of it like a single column in Excel.

## Creating Series from List
A Series is a 1D labeled array.

When we pass a Python list to pd.Series(), Pandas automatically assigns an index starting from 0.

```python
s1=pd.Series([10,20,30])
s1
```

We can access values using the index number.

```python
s1[2]
```

## Creating Series with Custom Index
We can access values using the index number.

```python
s=pd.Series([101,104,108], index=["a","b","c"])
s
```

```python
s=pd.Series([101,104,108], index=["a","b","c"])
s.loc["c"]=5
s.loc["a"]=200
s
```

## Creating Series from Dictionary

```python
dict={"Ahmad":90, "Ali": 85, "Zain": 77}
s=pd.Series(dict)
s
```

## Creating Series from NumPy Array

```python
arr=np.array([10,20,30,40,50,60,70])
s=pd.Series(arr)
s
```

## Creating Series using range()

```python
s=pd.Series(range(5,10))
s
# s=pd.Series(np.linspace(5,10,10))
# s
```

## Automatic Alignment

Pandas aligns data using index labels.

```python
s1=pd.Series([10,20,30], index=["a", "b", "c"])
s2=pd.Series([50,60,70], index=["d", "e", "f"])
s1+s2
```

## Handling Missing Values

```python
s1= pd.Series([1,np.nan, 3])
s1
```

```python
s1.dropna()  #dropping missing values
```

```python
s1.fillna(50)  #changing the nan value to some value
```

```python
s1.isnull()
```

```python
s1.notnull()
```

## Boolean filtering

```python
s1=pd.Series([10,20,30,40,50])
s1[s1>30]
```

```python
s1.unique()
```
