# Pandas DataFrame 

This section covers everything about Pandas DataFrames.

A DataFrame is a 2D labeled data structure with rows and columns.

```python
import pandas as pd
import numpy as np
```

## What is a DataFrame?

A DataFrame:
- is two-dimensional (rows and columns)
- has labeled columns and index
- can store different data types

Think of it like an Excel sheet.

```python
data={
    'Name': ['Ahmad', 'Ali', 'Zain', 'John'],
    'Age' : [20, 18, 21, 22],
    'City' : ['Lahore', 'Karachi', 'Faisalabad', 'Newyork'],
    'Field':['BSAI', 'BSCS', 'BSDS', 'BSCS']
}

df=pd.DataFrame(data)
df
```

## Creating DataFrame from List of Dictionaries

```python
data= [
    {'Name': 'Ahmad', 'Age': 20, 'Roll number': 37},
    {'Name': 'Ali', 'Age': 19, 'Roll number': 40}
]
df=pd.DataFrame(data)
df
```

## Custom Index

```python
df.index=['Student 1', 'Student 2']
print(df)
```

## DataFrame Structure

```python
print(df.shape)
print(df.columns)
print(df.index)
print(df.dtypes)
```

```python
df.head()
```

## Selecting Columns

```python
df['Name']         # Series
df[['Name', 'Age']]    # DataFrame
```

## Selecting Rows

```python
df.loc['Student 1'] #by label
```

```python
df.drop('Age', axis=1)
```
